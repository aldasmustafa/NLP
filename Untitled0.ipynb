```python
import nltk
import pandas as pd
from nltk.stem import PorterStemmer # Ya da SnowballStemmer

# nltk data'larn indirin (eer daha nce indirmediyseniz)
nltk.download('punkt')

# DataFrame'inizi ykleyin (lemmatized_data.csv veya hangi dosyay kullanyorsanz)
df = pd.read_csv('/content/drive/MyDrive/lemmatized_data.csv')

# PorterStemmer nesnesi oluturun
stemmer = PorterStemmer()

# Stemming fonksiyonu
def stem_tokens(tokens):
    # tokens_lemmatized stunundaki string'i listeye evir
    token_list = ast.literal_eval(tokens)
    return [stemmer.stem(token) for token in token_list]

# 'tokens_stemmed' stununu oluturun ve stemming fonksiyonunu uygulayn
df['tokens_stemmed'] = df['tokens_lemmatized'].apply(stem_tokens)

# Sadece 'tokens_stemmed' stununu ieren yeni bir DataFrame oluturun
stemmed_df = df[['tokens_stemmed']]

# Sonucu CSV dosyasna kaydedin
stemmed_df.to_csv('stemmed_data.csv', index=False)

print("Stemming ilemi tamamland ve 'stemmed_data.csv' dosyasna kaydedildi.")
```

```python
import nltk
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Gerekli nltk verilerini indirin
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Excel dosyasn okuyun
df = pd.read_excel('/content/drive/MyDrive/bank.xlsx')

# TRANSACTION DETAILS stununu sein ve bo deerleri atn
df = df[['TRANSACTION DETAILS']].dropna()
df.columns = ['aciklama']  # Stun adn deitirin

# n ileme admlar
def preprocess_text(text):
    text = str(text).lower()  # Kk harfe dntr
    text = re.sub(r'[^\w\s]', '', text)  # Noktalama iaretlerini kaldr
    tokens = word_tokenize(text)  # Tokenletir
    stop_words = set(stopwords.words('english'))  # Stop word'leri tanmla
    tokens = [w for w in tokens if w not in stop_words]  # Stop word'leri kaldr
    return tokens

df['tokens'] = df['aciklama'].apply(preprocess_text)

# Lemmatization
lemmatizer = WordNetLemmatizer()
def lemmatize_tokens(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]

df['tokens_lemmatized'] = df['tokens'].apply(lemmatize_tokens)

# Sonular kaydedin
df[['tokens_lemmatized']].to_csv('lemmatized_data.csv', index=False)

print("n ileme ve Lemmatization tamamland. Sonular 'lemmatized_data.csv' dosyasna kaydedildi.")
```
