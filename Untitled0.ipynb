```python
def temizle_ve_tokenle(metin):
    metin = str(metin).lower()  # Küçük harf
    metin = re.sub(r'<.*?>', '', metin)  # HTML etiketlerini kaldır
    metin = re.sub(r'[^\w\s]', '', metin)  # Noktalama işaretlerini kaldır
    metin = re.sub(r'\d+', '', metin)  # Sayıları kaldır
    return metin.split()
```

```python
import nltk
import pandas as pd
from nltk.stem import PorterStemmer # Ya da SnowballStemmer

# nltk data'larını indirin (eğer daha önce indirmediyseniz)
nltk.download('punkt')

# DataFrame'inizi yükleyin (lemmatized_data.csv veya hangi dosyayı kullanıyorsanız)
df = pd.read_csv('/content/drive/MyDrive/lemmatized_data.csv')

# PorterStemmer nesnesi oluşturun
stemmer = PorterStemmer()

# Stemming fonksiyonu
def stem_tokens(tokens):
    # tokens_lemmatized sütunundaki string'i listeye çevir
    token_list = ast.literal_eval(tokens)
    return [stemmer.stem(token) for token in token_list]

# 'tokens_stemmed' sütununu oluşturun ve stemming fonksiyonunu uygulayın
df['tokens_stemmed'] = df['tokens_lemmatized'].apply(stem_tokens)

# Sadece 'tokens_stemmed' sütununu içeren yeni bir DataFrame oluşturun
stemmed_df = df[['tokens_stemmed']]

# Sonucu CSV dosyasına kaydedin
stemmed_df.to_csv('stemmed_data.csv', index=False)

print("Stemming işlemi tamamlandı ve 'stemmed_data.csv' dosyasına kaydedildi.")
```

```python
import nltk
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Gerekli nltk verilerini indirin
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Excel dosyasını okuyun
df = pd.read_excel('/content/drive/MyDrive/bank.xlsx')

# TRANSACTION DETAILS sütununu seçin ve boş değerleri atın
df = df[['TRANSACTION DETAILS']].dropna()
df.columns = ['aciklama']  # Sütun adını değiştirin

# Ön işleme adımları
def preprocess_text(text):
    text = str(text).lower()  # Küçük harfe dönüştür
    text = re.sub(r'[^\w\s]', '', text)  # Noktalama işaretlerini kaldır
    tokens = word_tokenize(text)  # Tokenleştir
    stop_words = set(stopwords.words('english'))  # Stop word'leri tanımla
    tokens = [w for w in tokens if w not in stop_words]  # Stop word'leri kaldır
    return tokens

df['tokens'] = df['aciklama'].apply(preprocess_text)

# Lemmatization
lemmatizer = WordNetLemmatizer()
def lemmatize_tokens(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]

df['tokens_lemmatized'] = df['tokens'].apply(lemmatize_tokens)

# Sonuçları kaydedin
df[['tokens_lemmatized']].to_csv('lemmatized_data.csv', index=False)

print("Ön işleme ve Lemmatization tamamlandı. Sonuçlar 'lemmatized_data.csv' dosyasına kaydedildi.")
```
